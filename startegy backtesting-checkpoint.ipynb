```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
crypto_processed_data = pd.read_csv("crypto_processed_data.csv")
```

```python
crypto_processed_data.head()
```

```python
crypto_processed_data['signal'] = np.where(
    (crypto_processed_data['rsi'] < 30) & (crypto_processed_data['sentiment_score'] > 0),
    1,   # BUY
    np.where(
        (crypto_processed_data['rsi'] > 70) & (crypto_processed_data['sentiment_score'] < 0),
        -1,  # SELL
        0    # HOLD
    )
)
```

```python
crypto_processed_data['strategy_return'] = (
    crypto_processed_data['signal'].shift(1) * crypto_processed_data['returns']
)
```

```python
crypto_processed_data['market_cum_return'] = (
    1 + crypto_processed_data['returns']
).cumprod()

crypto_processed_data['strategy_cum_return'] = (
    1 + crypto_processed_data['strategy_return']
).cumprod()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(crypto_processed_data['date'], crypto_processed_data['market_cum_return'], label='Market Return')
plt.plot(crypto_processed_data['date'], crypto_processed_data['strategy_cum_return'], label='Strategy Return')
plt.legend()
plt.title("Strategy vs Market Performance")
plt.show()
```

```python
total_strategy_return = crypto_processed_data['strategy_cum_return'].iloc[-1] - 1
total_market_return = crypto_processed_data['market_cum_return'].iloc[-1] - 1

total_strategy_return, total_market_return
```

```python
sharpe_ratio = (
    crypto_processed_data['strategy_return'].mean() /
    crypto_processed_data['strategy_return'].std()
) * np.sqrt(252)

sharpe_ratio
```

```python
rolling_max = crypto_processed_data['strategy_cum_return'].cummax()
drawdown = crypto_processed_data['strategy_cum_return'] / rolling_max - 1
max_drawdown = drawdown.min()

max_drawdown
```

```python
backtest_summary = pd.DataFrame({
    'Metric': ['Total Return', 'Sharpe Ratio', 'Max Drawdown'],
    'Strategy': [total_strategy_return, sharpe_ratio, max_drawdown],
    'Market': [total_market_return, None, None]
})

backtest_summary
```

```python
crypto_processed_data.to_csv("strategy_backtest_data.csv", index=False)
backtest_summary.to_csv("backtest_summary.csv", index=False)
```
