### 1.1 To Import Library

```python
import requests
import pandas as pd
import numpy as np
```

### 1.2 To get OHLCV data into Binance API 

```python
url = "https://api.binance.com/api/v3/klines"

params = {
    "symbol": "BTCUSDT",
    "interval": "1d",
    "limit": 500
}

data = requests.get(url, params=params).json()

df = pd.DataFrame(data, columns=[
    'open_time','open','high','low','close','volume',
    'close_time','qav','num_trades',
    'taker_base_vol','taker_quote_vol','ignore'
])

df['date'] = pd.to_datetime(df['open_time'], unit='ms')

df = df[['date','open','high','low','close','volume']]
df[['open','high','low','close','volume']] = df[['open','high','low','close','volume']].astype(float)

df.set_index('date', inplace=True)
```

### 1.3 Returns & Log Returns

```python
df['returns'] = df ['close'].pct_change()
df['log_returns']=np.log(df['close']/df['close'].shift(1))
```

### 1.4 Volatility (Rolling 20 days)

```python
df['volatility'] = df['log_returns'].rolling(20).std()
```

### 1.5  Moving Averages

```python
df['ma_20'] = df['close'].rolling(20).mean()
df['ma_50'] = df['close'].rolling(50).mean()
```

### 1.6 RSI (Relative Strength Index)

```python
delta = df['close'].diff()

gain = np.where(delta > 0, delta, 0)
loss = np.where(delta < 0, -delta, 0)

avg_gain = pd.Series(gain, index=df.index).rolling(14).mean()
avg_loss = pd.Series(loss, index=df.index).rolling(14).mean()

rs = avg_gain / avg_loss
df['rsi'] = 100 - (100 / (1 + rs))
```

### 1.7 Sentiment Score

```python
df['sentiment_score'] = np.where(df['returns'] > 0, 1, -1)
```

### 1.8 Market Regime (Bull / Bear / Sideways)

```python
conditions = [
    (df['ma_20'] > df['ma_50']),
    (df['ma_20'] < df['ma_50'])
]

choices = ['Bull', 'Bear']

df['regime'] = np.select(conditions, choices, default='Sideways')
```

### 1.9 Final Clean Data

```python
df = df.dropna()
print(df.head())
```

### 1.10 CSV Export (Power BI ready)

```python
df.to_csv("crypto_full_feature_dataset.csv")
```

## Task 3 Time Series Forcasting Models

### 3.1 Remove Nan  

```python
df_model = df.dropna().copy()
df_model.shape
```

### 3.2 Feature Selection

```python
features = [
    'returns',
    'log_returns',
    'volatility',
    'ma_20',
    'ma_50',
    'rsi',
    'sentiment_score'
]

X = df_model[features]
y = df_model['regime']   # Bull / Bear
```

### 3.3 Regime Encoding (Bull / Bear  1 / 0)

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Bull = 1 , Bear = 0
```

### 3.4 Train - Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded,
    test_size=0.2,
    shuffle=False   # time-series rule
)
```

### 3.5 Model Training (Logistic Regression)

```python

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

### 3.6 Prediction &  evaluation

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

### 3.7 Result Interpretation

```python
df_model['predicted_regime'] = le.inverse_transform(
    model.predict(X)
)

df_model[['date', 'regime', 'predicted_regime']].tail()
```

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
```

```python
# Data preparation

df = pd.read_csv("crypto_processed_data.csv")

df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

close_series = df['close']
```

```python
# Train test split

train = close_series[:-30]
test = close_series[-30:]
```

```python
# ARIMA Model Build
model = ARIMA(train, order=(5,1,0))
model_fit = model.fit()
```

```python
# libraries

import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
# Scaling Data
scaler = MinMaxScaler()
scaled_close = scaler.fit_transform(df[['close']])
```

```python
# Create Seqence
X, y = [], []

for i in range(60, len(scaled_close)):
    X.append(scaled_close[i-60:i])
    y.append(scaled_close[i])

X, y = np.array(X), np.array(y)
```

```python
# Train-Test Split

split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
# Lstm model

model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(LSTM(50))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
# Train Model

model.fit(X_train, y_train, epochs=10, batch_size=32)
```

```python
# prediction

predicted = model.predict(X_test)
predicted = scaler.inverse_transform(predicted)
```
